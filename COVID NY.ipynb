```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as dates

df = pd.read_csv('NY_Counties.csv')
df.tail()
```

Starting to collect county data statewide on 3/26

```python
#Remove commas,NaN, set to int
df = df.replace(',','', regex=True)
dff = df.fillna(0)

dff = dff.reset_index()
dff = dff.drop(columns='Date')
dff = dff.astype(int) 

#Transpose
df = dff
#dft = df.T

df.rename(index=str, columns={"New York City": "nyc"},inplace=True)

#Calculate diffrence
diff = dff.diff()
```

```python
#Plot state cases
plt.figure(figsize=(10,9))

plt.plot(df.index,df.tot_pos,label='Total Cases')
plt.scatter(df.index, df.tot_pos,s=40)

plt.plot(df.index,diff.tot_pos,'--r',label='Daily Cases')
plt.scatter(df.index, diff.tot_pos,s=40)
plt.fill_between(df.index,diff.tot_pos,color='r',alpha=0.2)

plt.title('COVID-19: New York')
plt.ylabel('Positive Cases State')
plt.xlabel('Date')

leg = plt.legend();

plt.show()
```

```python
#Plot top counties

plt.figure(figsize=(10,9))

plt.plot(df.index,df.nyc,label='New York City')
plt.scatter(df.index,df.nyc)

plt.plot(df.index,df.Suffolk,label='Suffolk')
plt.scatter(df.index,df.Suffolk)

plt.plot(df.index,df.Nassau,label='Nassau')
plt.scatter(df.index,df.Nassau)

plt.plot(df.index,df.Westchester,label='Westchester')
plt.scatter(df.index,df.Westchester)

plt.title('COVID-19: New York Counties')
plt.ylabel('Positive Cases')
plt.xlabel('Date')
leg = plt.legend();
plt.show()
```
